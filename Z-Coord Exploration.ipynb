```python
import sys
import os

sys.path.append('data')
import coordKeysZ as keys
```

#Extract Data

```python
data = []
with open('data/squatData10.txt') as f:
    for line in f:
        data.append([float(x.replace('\r\n','')) for x in line.split(',')])
```

#Investigate Z Coords

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.DataFrame(data)
df.columns = keys.columns
```

##Squater at the beginning of rep

We're plotting y coordinates (y-axis) vs. z coordinates (x-axis)

```python
x1 = [df.loc[1,z[0:-1] +'Z'] for z in df.columns]
y1 = [-df.loc[1,y[0:-1] +'Y'] for y in df.columns]
plt.plot(x1, y1,linestyle='none',marker='o')
```

##Squater some time into Rep

```python
x2 = [df.loc[100,z[0:-1] +'Z'] for z in df.columns]
y2 = [-df.loc[100,y[0:-1] +'Y'] for y in df.columns]
plt.plot(x2, y2,linestyle='none',marker='o')
```

```python
plt.plot(x1,y1,linestyle='none',marker='o',color='b',label='Starting Position')
plt.plot(x2,y2,linestyle='none',marker='o',color='r',label='Ending Position')
plt.legend(loc='upper left')
```

Notice how there's a clear shift in the z-coordinates. The hips move farther from the kinect and the arms move closer.


##Squatter's feet z-coords throughout squat

```python
import numpy as np

feet_coords = np.concatenate([df.get('FootLeftZ'),df.get('FootRightZ')],axis=1)
plt.plot(feet_coords)
plt.axvline(df.get('FootRightZ').size,color='r')
```

Left side are z-coords for left foot, and right side are z-coords for right foot.
